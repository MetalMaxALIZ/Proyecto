```python
import pickle
import pandas as pd
from pathlib import Path
import json

# Cargar pickle
with open('checkpoints/apps_dict-ckpt-fin.p', 'rb') as f:
    apps_dict = pickle.load(f)

# Convertir a lista de diccionarios aplanados
games_list = []
for appid, game_data in apps_dict.items():
    game = {
        'appid': appid,
        'name': game_data.get('name'),
        'type': game_data.get('type'),
        'is_free': game_data.get('is_free'),
        'genres': ', '.join([g['description'] for g in game_data.get('genres', [])]),
        'price': game_data.get('price_overview', {}).get('final', 0) / 100 if game_data.get('price_overview') else 0,
        'metacritic_score': game_data.get('metacritic', {}).get('score') if game_data.get('metacritic') else None,
        'header_image': game_data.get('header_image'),
        'recomendations': game_data.get('recommendations', {}).get('total', 0) if game_data.get('recommendations') else 0,

    }
    games_list.append(game)

df = pd.DataFrame(games_list)
df.to_csv('steam_games.csv', index=False, encoding='utf-8')
print(f"✓ Guardados {len(df)} juegos en steam_games.csv")
```
